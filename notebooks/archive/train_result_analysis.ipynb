```python
import pandas as pd
import numpy as np
```

```python
linear_data = pd.read_csv('../results/train/linear_PolicyModel/full_data/base_case/train_results.csv', index_col=0)
non_linear_data = pd.read_csv('../results/train/non_linear_PolicyModel/full_data/base_case/train_results.csv', index_col=0)
only_DA_data = pd.read_csv('../results/train/only_DA_non_linear_PolicyModel/full_data/base_case/train_results.csv', index_col=0)


interesting_columns = ['date', 'energy_RE', 'consumption_electrolyzer', 'hydrogen_output', 'volume_DA', 'lambda_DA_RE', 'DA_revenue', 'hydrogen_revenue', 'volume_IM', 'lambda_IM', 'IM_revenue']
linear_data[interesting_columns]
```

```python
linear_data.columns
```

```python
non_linear_data = pd.read_csv('../results/train/non_linear_PolicyModel/full_data/base_case/train_results.csv', index_col=0)

non_linear_data[interesting_columns]
```

```python
only_DA_data = pd.read_csv('../results/train/only_DA_non_linear_PolicyModel/full_data/base_case/train_results.csv', index_col=0)

only_DA_data[interesting_columns]
```

```python
hindsight_data = pd.read_csv('../results/train/HindsightModel/full_data/base_case/train_results.csv', index_col=0)
hindsight_data[interesting_columns]
```

```python
comparison_table = pd.DataFrame(index=linear_data.index, columns=['linear', 'non_linear', 'only_DA', 'hindsight'])
comparison_table['linear'] = linear_data['consumption_electrolyzer']
comparison_table['non_linear'] = non_linear_data['consumption_electrolyzer']
comparison_table['only_DA'] = only_DA_data['consumption_electrolyzer']
comparison_table['hindsight'] = hindsight_data['consumption_electrolyzer']
comparison_table['lambda_DA_RE'] = hindsight_data['lambda_DA_RE']

comparison_table['min_price'] = np.minimum(hindsight_data['lambda_DA_RE'], hindsight_data['lambda_IM'])

comparison_table['linear_H_cost'] = comparison_table['linear'] * comparison_table['min_price']
comparison_table['non_linear_H_cost'] = comparison_table['non_linear'] * comparison_table['min_price']
comparison_table['only_DA_H_cost'] = comparison_table['only_DA'] * comparison_table['lambda_DA_RE']
comparison_table['hindsight_H_cost'] = comparison_table['hindsight'] * comparison_table['min_price']

comparison_table['linear_H_revenue'] = linear_data['hydrogen_revenue']
comparison_table['non_linear_H_revenue'] = non_linear_data['hydrogen_revenue']
comparison_table['only_DA_H_revenue'] = only_DA_data['hydrogen_revenue']
comparison_table['hindsight_H_revenue'] = hindsight_data['hydrogen_revenue']

comparison_table['linear_H_profit'] = comparison_table['linear_H_revenue'] - comparison_table['linear_H_cost']
comparison_table['non_linear_H_profit'] = comparison_table['non_linear_H_revenue'] - comparison_table['non_linear_H_cost']
comparison_table['only_DA_H_profit'] = comparison_table['only_DA_H_revenue'] - comparison_table['only_DA_H_cost']
comparison_table['hindsight_H_profit'] = comparison_table['hindsight_H_revenue'] - comparison_table['hindsight_H_cost']

comparison_table['linear_difference'] = abs(comparison_table['linear'] - comparison_table['hindsight'])
comparison_table['non_linear_difference'] = abs(comparison_table['non_linear'] - comparison_table['hindsight'])
comparison_table['only_DA_difference'] = abs(comparison_table['only_DA'] - comparison_table['hindsight'])

comparison_table.sum()
```
